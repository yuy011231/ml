```python
import sys
from pathlib import Path
import pandas as pd
import lightgbm as lgbm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris

src_path = Path.cwd().parent
sys.path.append(str(src_path))

from util.reader import CSVReader
from util.tools import correlation_heatmap
```

```python
train_reader = CSVReader(Path("data/train.csv"))
test_reader = CSVReader(Path("data/test.csv"))
```

```python
passenger_Id = "PassengerId"
sex = "Sex"
embarked = "Embarked"
name = "Name"
cabin = "Cabin"
ticket = "Ticket"
survived = "Survived"
```

```python
train = train_reader.read()
test = test_reader.read()
train = train.set_index(passenger_Id)
test = test.set_index(passenger_Id)

check_df = pd.concat([train, test], axis=0, sort=False)
# print(check_df)
check_df.isnull().sum()
check_df.info()

check_df[embarked].unique()
```

```python
def cleaning(df: pd.DataFrame) -> pd.DataFrame:
    rtn_df = df
    # ラベルエンコーディング
    rtn_df[sex] = df[sex].map({"female": 1, "male": 2})
    rtn_df[embarked] = df[embarked].map({"NaN": 0, "S": 1, "C": 2, "Q": 3})

    # 欠損値処理
    rtn_df[embarked] = rtn_df[embarked].fillna(rtn_df[embarked].median())
    
    # 使わないカラムの削除
    rtn_df = rtn_df.drop(columns=[name, cabin, ticket])
    return rtn_df
```

```python
train = cleaning(train)
# print(train)
correlation_heatmap(train)
```

```python
# データ分割
X = train.drop(columns=[survived])
y = train[survived]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# モデル
model = lgbm.LGBMClassifier(objective='binary', n_estimators=100)
model.fit(X_train, y_train)

predictions = model.predict(X_test)
print("\n予測結果（最初の5件）:", predictions[:5])
print("正解（最初の5件）:  ", y_test.values[:5])

accuracy = accuracy_score(y_test, predictions)
print(f"\nモデルの正解率: {accuracy:.4f}")
```
